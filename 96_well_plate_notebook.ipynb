```python
"""
Notebook used to read a xlsx file containing sheets of 96 well plates from the TECAN
"""

# Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statannotations.Annotator import Annotator


# Apply theme
sns.set_theme(style="whitegrid")

# Import csv file
input_file = "./20221206 TQ MB ATP E3.xlsx"
df = pd.read_excel(input_file)

# Print dataframe
df
```

```python
# input: excel sheet as df
# output: single column A1-H12 as df

# Stacking Method
##############################################
# isolate 96 well plate data
# stack all
#isolate values from row 0-95 in column 2
def plate_2_df_column(df):
    df1=df.iloc[22:30,1:13]
    df2 = df1.stack().reset_index()
    df3=df2.iloc[0:96,2]
    return df3
```

```python
# Iterate through sheets Method
##############################################
# import xlsx with all sheets
# itertate through all of sheets
# load sheet to df
# transform 96 well plate to single column
# add column to master df with sheet name as column name
# add Group column and annotate
# export master_df as csv file

input_file = './20221206 TQ MB ATP E3.xlsx'

xls = pd.ExcelFile(input_file)
sheet_list = xls.sheet_names
master_df = pd.DataFrame()

for i in range(len(sheet_list)):
    current_df = pd.read_excel(xls, sheet_list[i])
    df_to_add = plate_2_df_column(current_df)
    master_df.insert(loc=i, column=sheet_list[i], value=df_to_add,)

Group_list=[]
Groups=["Control1", "TQ 10uM", "TQ 2uM", "MB 800nM", "MB 100nM", "TQ&MB hi", "TQ&MB lo", "Control2"]
for i in range(8):
    for j in range(12):
        Group_list.append(Groups[i])
    
master_df.insert(len(sheet_list),column="Group",value=Group_list)
master_df.to_csv(f'{input_file} CONVERSION',index=False)
```
